## Pandas Dataframes

```python
import numpy as np
import pandas as pd
```

```python
names = ["john", "shuri", "erik", "simon", "daphane"]
heights = ["72", "63", "74", "73", "74"]
weights = ["180", "150", "215", "200", "183"]
```

```python
df = pd.DataFrame({"names": names, "heights": heights, "weights": weights})
df
```

```python
df = pd.DataFrame({"names": names, "heights": heights, "weights": weights}, index=names)
df
```

### Basic Indexing

```python
df["heights"]
```

```python
df["heights"][0:2]
```

```python
i = "weights"
df[i]
```

```python
df.weights
```

```python
df.weights[::-1]
```

```python
# Directly dealing with names
# Inclusive of stopping point; Because we don't have a numerical based indexing
df.loc["john":"erik"]
```

```python
df_num = pd.DataFrame({"names": names, "heights": heights, "weights": weights})
display(df_num)

# Even if the index si numerical, the ending index is included.
df_num.loc[0:3]
```

```python
# iloc operates on he numerical index of the row
# df's index is also names only.
df.iloc[0::2]
```

### Basic Math Operations

```python
mcd.describe()
```

```python
mcd["Calories"].mean()
```

```python
mcd.sort_values("Calories")
```

```python
# Stringing together methods
display(mcd[mcd["Category"] == "Breakfast"].sort_values("Calories", ascending=False).head(1))

# The direct [0] at the end does not work. This is because the result's index doesn't have 0
# The square bracket indexing works as loc instead of iloc.
mcd[mcd["Category"] == "Breakfast"].sort_values("Calories", ascending=False)
```

```python
mcd["john_index"] = 12*(mcd["Protein"] + mcd["Dietary Fiber"])/mcd["Calories"]
mcd.head(2)
```

```python
# Notice that there are 12 values which aren't correct
mcd["john_index"].info()
```

```python
mcd["john_index"].unique()
```

```python
mcd[~mcd["john_index"].notnull()]
```

```python
mcd["john_index"].fillna(mcd["john_index"].mean()).info()
```

```python
mcd["john_index"].replace(np.inf, mcd["john_index"].mean())
```

```python
mcd.groupby("Category").mean()
```

```python
cal_summary = mcd.groupby("Category").agg(mean_cal = ("Calories", "mean")).reset_index()
cal_summary
```

```python
cat_mcd = mcd[(mcd["Category"] == "Beef & Pork") | (mcd["Category"] == "Breakfast") | (mcd["Category"] == "Chicken & Fish")]
cat_mcd
```

```python
# Map each category to a color
colormap = {"Beef & Pork": "red", "Breakfast": "gold", "Chicken & Fish": "teal"}
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.scatter(cat_mcd["Calories"], cat_mcd["Protein"], s=cat_mcd["Total Fat"]*4, c=cat_mcd["Category"].map(colormap), alpha=0.5)
plt.xlabel("Calories")
plt.ylabel("Proteins")
plt.grid()
plt.show()
```
